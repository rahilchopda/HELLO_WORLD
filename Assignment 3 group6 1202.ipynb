```python
# How many games on this list were launched between 2000-2005? 

# Import pandas
import pandas as pd
from sqlalchemy import create_engine
import pymysql
```

```python
# read a simple query into DataFrame
df = pd.read_sql("SELECT * FROM vgsales", conn)
```

```python
#Average sales before 2005
average_sales_pre = pd.read_sql("SELECT 'before 2005' as Period, AVG(global_sales) as Average_Sales from vgsales where Year_of_Release < 2005",conn)
```

```python
average_sales_pre
```

```python
average_sales_post = pd.read_sql("SELECT 'after 2005' as Period, AVG(global_sales) as Average_Sales from vgsales where Year_of_Release > 2005",conn)
```

```python
average_sales_post
```

```python
#
average_sales = pd.read_sql("SELECT 'before 2005' as Period, AVG(global_sales) as Average_Sales from vgsales where Year_of_Release < 2005 union SELECT 'after 2005' as Period, AVG(global_sales) as Average_Sales from vgsales where Year_of_Release > 2005",conn)
```

```python
average_sales
```

```python
#Create a new column that labels records before 2005 as 'pre-2005' and after 2005 as 'post-2005'
def find_default_value(col_val):
    if col_val < 2005:
        return 'pre-2005'
    else:
        return 'post-2005'
    
df['Year_Category'] = df['Year_of_Release'].apply(find_default_value)

df
```
